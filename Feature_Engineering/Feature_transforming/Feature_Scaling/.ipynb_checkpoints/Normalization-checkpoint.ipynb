```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('wine_data.csv',header=None,usecols=[0,1,2])
df.columns=['Class','Alcohol','Malic_acid']
df
```

```python
df.sample(5)
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic_acid'])
```

```python
color_dict={1:'red',2:'green',3:'yellow'}
sns.scatterplot(x='Alcohol',y='Malic_acid',hue='Class',data=df,palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(['Class'],axis=1),
                                                df['Class'], 
                                               test_size=0.2, 
                                               random_state=42
                                              )
X_test.shape,X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler() 
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled=pd.DataFrame(X_test_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```
