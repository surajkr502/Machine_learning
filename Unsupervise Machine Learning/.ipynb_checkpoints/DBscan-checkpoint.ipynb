```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.cluster import DBSCAN
```

```python
x=np.array([[1,2],[2,2],[5,7],[8,8],[25,80]])
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
db=DBSCAN(eps=1,min_samples=3) #bcz 80 contaions only 1 sample ,so it is outlier(-1)
db.fit(x)
db.labels_
```
