```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')
```

```python
dt=pd.read_csv("D:/Ad_py_lib.csv/ml_csv/Kmeanssalry.csv")
```

```python
dt
```

```python
scaler=MinMaxScaler() 
scaler.fit(dt[['Age']])
dt['Age']=scaler.transform(dt[['Age']])
scaler.fit(dt[['Income']])
dt['Income']=scaler.transform(dt[['Income']])
```

```python
sse=[] 
for i in range(1,10):
    sk=KMeans(n_clusters=i)
    sk.fit_predict(dt)
    sse.append(sk.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Square Error')
plt.plot(range(1,10),sse,marker='D')
```
