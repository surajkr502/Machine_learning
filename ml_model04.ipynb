```python
x=df.drop(columns=['price'],axis=1)
y=df['price']
```

```python
x
```

```python
x_one_encode=pd.get_dummies(x,columns=['model','transmission','fuelType'],drop_first=True)
```

```python
x_one_encode=x_one_encode.astype(int)
```

```python
x_one_encode
```

```python
from sklearn.preprocessing import LabelEncoder 
columns=['model','transmission','fuelType']
xlabel=x.copy()  #to make safe copy 
label_encoders={} 

# xlabel=x
for col in columns:
    le=LabelEncoder()
    xlabel[col]=le.fit_transform(xlabel[col].astype(str)) #To convert the string in case of nulls
    label_encoders[col]=le
```

```python
xlabel
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
numeric_col=['year','mileage','tax','mpg','engineSize']
x_one_encode[numeric_col]=scaler.fit_transform(x_one_encode[numeric_col])
```

```python
x_one_encode
```

```python
xlabel.columns
```

```python
xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']]=scaler.fit_transform(xlabel[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
xlabel
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x_one_encode, y, test_size=0.20, random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2=r2_score(y_test,y_pred)
r2
```

```python
n=X_test.shape[0]
p=X_test.shape[1]
adjusted_r2=1-((1-r2)*(n-1))/(n-p-1)
print("adjusted r2 score:",adjusted_r2)
```
