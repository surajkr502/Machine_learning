```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings("ignore")
```

```python
df=sns.load_dataset('titanic')
```

```python
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.value_counts('embarked')
```

```python
df['embarked'].isnull().sum()  #There are two null value in embarked.
```

```python
df['age'].isnull().sum()    #There are 177 null value in age.
```

```python
df.drop(['deck', 'embark_town','class','alive','who', 'adult_male'],axis=1,inplace=True)
```

```python
df
```

```python
# fill the null value by the means value function 
df['age'].fillna(df['age'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['embarked'],inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# label encoding 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.head(5)
```

```python
df['sex']=le.fit_transform(df['sex'])
df['embarked']=le.fit_transform(df['embarked'])
```

```python
df=df.astype(int)
```

```python
df.head(5)
```

```python
x=df.drop('survived',axis=1)
y=df['survived']
```

### LogisticRegression

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression() 
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# confussion matrix
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

### KNeighborsClassifier

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler() 
```

```python
X_train_scaled=scaler.fit_transform(X_train) 
X_test_scaled=scaler.fit_transform(X_test)
# X_train_scaled

```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_scaled,y_train)
```

```python
y_pred_knn=knn_model.predict(X_test_scaled)
```

```python
y_pred_knn
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred_knn)
```

```python
confusion_matrix(y_test,y_pred_knn)
```

```python
print(classification_report(y_test,y_pred_knn))
```
