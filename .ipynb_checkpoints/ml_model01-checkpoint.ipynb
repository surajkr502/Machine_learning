### Data preprocessing

```python
df_encode=pd.get_dummies(df,drop_first=True)
```

```python
df_encode
```

```python
df_encode=df_encode.astype(int)
```

```python
df_encode
```

```python
from sklearn.preprocessing import StandardScaler
numerical_cols=['Age','RestingBP','Cholesterol',
                'MaxHR','Oldpeak']
scaler=StandardScaler() 
df_encode[numerical_cols]=scaler.fit_transform(df_encode[numerical_cols])
df_encode.head()
```

```python
df_encode=df_encode.astype(int)
```

```python
df_encode
```

```python
df_encode.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,f1_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
X=df_encode.drop('HeartDisease',axis=1)
y=df_encode['HeartDisease']
```

```python
# shift+Tab
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

```python
models={
    'Logistic Regression':LogisticRegression(),
    'KNN':KNeighborsClassifier(),
    'Neive Byeas':GaussianNB(),
    'decision tree':DecisionTreeClassifier(), 
    'SVM (RBF Kernel)': SVC(probability=True)
}
```

```python
result=[] 
```

```python
for name,models_01 in models.items(): 
    models_01.fit(X_train_scaled,y_train)
    y_pred=models_01.predict(X_test_scaled)
    acc=accuracy_score(y_test,y_pred)
    f1=f1_score(y_test,y_pred)
    result.append({
    'Model':name,
    'Accuracy':round(acc,4), 
    'f1 score': round(f1,4)
    })
    
```

```python
result
```

```python
import joblib
joblib.dump(models['KNN'],'model_01_KNN_heart.pkl')
joblib.dump(scaler,'scaler.pkl')
joblib.dump(X.columns.tolist(),'columns.pkl')
```
